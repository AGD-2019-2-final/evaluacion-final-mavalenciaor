```python
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -rm data.tsv .
!hadoop fs -put data.tsv .
!hadoop fs -ls
```

```python
%%pig
records = LOAD 'data.tsv' AS (c1:CHARARRAY, c2:BAG{t:tuple(p:CHARARRAY)}, c3:MAP[]);
DUMP records;
```

```python
%%pig
conteo = FOREACH records GENERATE c1, COUNT(c2) AS c2_tam, SIZE(c3) AS c3_tam;
ordered_data = ORDER conteo BY c1 ASC, c2_tam ASC, c3_tam ASC;
DUMP ordered_data;
```

```python
%%pig
STORE ordered_data INTO 'output';
```

```python
%%pig
fs -get output/ .
```
