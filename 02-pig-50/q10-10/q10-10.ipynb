```python
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -rm data.csv .
!hadoop fs -put data.csv .
!hadoop fs -ls
```

```python
%%pig
u = LOAD 'data.csv' USING PigStorage(',') 
    AS (id:int, 
        firstname:CHARARRAY, 
        surname:CHARARRAY, 
        birthday:CHARARRAY, 
        color:CHARARRAY, 
        quantity:INT);
DUMP u;
```

```python
%%pig
long_surname = FOREACH u GENERATE surname AS surname, SIZE(surname) AS sizeof;
sorted_l_surname = ORDER long_surname BY sizeof DESC;
limited_data = LIMIT sorted_l_surname 5; 
DUMP limited_data;
```

```python
%%pig
STORE limited_data INTO 'output';
```

```python
%%pig
fs -get output/ .
```

```python
!hadoop fs -rm output/*
!hadoop fs -rmdir output
```
