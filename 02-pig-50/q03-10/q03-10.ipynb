```python
%load_ext bigdata
%pig_start
%timeout 300
!hadoop fs -put *.tsv .
!hadoop fs -ls
```

```python
%%pig
lines = LOAD 'data.tsv' AS (let:CHARARRAY, dat:CHARARRAY, numb:INT);
DUMP lines;
```

```python
%%pig

numbers = FOREACH lines GENERATE numb;
DUMP numbers;
```

```python
%%pig
sorted_numbs = ORDER numbers BY numb ASC;
DUMP sorted_numbs;
```

```python
%%pig
limited_data = LIMIT sorted_numbs 5;
DUMP limited_data;
```

```python
%%pig
STORE limited_data INTO 'output';
```

```python
%%pig
fs -get output/ .
```

```python
!hadoop fs -rm output/*
!hadoop fs -rmdir output
```
